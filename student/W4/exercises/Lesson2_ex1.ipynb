## Task 3: Batch conversion


```python
pdf_urls = [
    "https://arxiv.org/pdf/2206.01062",  # Document layout analysis
    "https://arxiv.org/pdf/2203.01017",  # TableFormer paper
    "https://arxiv.org/pdf/2408.09869",  # Docling technical report
]
```

```python
pdf_opt = PdfPipelineOptions()

pdf_opt.do_ocr == True
pdf_opt.do_table_structure = True

pipeline_options.table_structure_options = TableStructureOptions(
    do_cell_matching=True
)
pipeline_options.accelerator_options = AcceleratorOptions(
        num_threads=4, device=AcceleratorDevice.AUTO
    )
```

```python
converter = DocumentConverter( #build the converter and add customizable options
        format_options={
            InputFormat.PDF: PdfFormatOption(pipeline_options=pdf_opt) # here it is more like (for PDF files --> use this option)
        }
    )
```

```python
results = converter.convert_all(pdf_urls,
                      raises_on_error=False) # don't crash if one fails
```

This loop i got from my VA

it only iterates through the iterable (results) and save the results to the path given

```python
from pathlib import Path
from docling.datamodel.base_models import ConversionStatus

start = time.time()

output_dir = Path("scratch")
output_dir.mkdir(parents=True, exist_ok=True)

success = 0
partial = 0
failed = 0

for result in results:
    filename = result.input.file.stem  # e.g. "2206.01062"

    if result.status == ConversionStatus.SUCCESS:
        result.document.save_as_markdown(output_dir / f"{filename}.md")
        result.document.save_as_json(output_dir / f"{filename}.json")
        success += 1
        print(f"✅ {filename}")

    elif result.status == ConversionStatus.PARTIAL_SUCCESS:
        # Still got something — save it, but log the issues
        result.document.save_as_markdown(output_dir / f"{filename}.md")
        partial += 1
        print(f"⚠️  {filename} — partial. Errors:")
        for err in result.errors:
            print(f"     {err.error_message}")

    else:  # FAILURE
        failed += 1
        print(f"❌ {filename} — failed")

elapsed = time.time() - start
print(f"\nDone in {elapsed:.2f}s — {success} ok, {partial} partial, {failed} failed")
```

```python
from docling.pipeline.simple_pipeline import SimplePipeline
from docling.pipeline.standard_pdf_pipeline import StandardPdfPipeline
from docling.backend.pypdfium2_backend import PyPdfiumDocumentBackend
from docling.datamodel.base_models import InputFormat
from docling.document_converter import (
    DocumentConverter,
    PdfFormatOption,
    WordFormatOption,
)

doc_converter = DocumentConverter(
        allowed_formats=[ #here we choose the allowed format and will ignore the dormats not specified when converting
            InputFormat.PDF,
            InputFormat.IMAGE,
            InputFormat.DOCX,
            InputFormat.HTML,
            InputFormat.PPTX,
            InputFormat.ASCIIDOC,
            InputFormat.CSV,
            InputFormat.MD,
            ],
        format_options={
            InputFormat.PDF: PdfFormatOption(
                pipeline_cls=StandardPdfPipeline, backend=PyPdfiumDocumentBackend
            ),
            InputFormat.DOCX: WordFormatOption(
                pipeline_cls=SimplePipeline)})
```

```python
from pathlib import Path

input_paths = [
    Path("README.md"),
    Path("wiki_duck.html"),
    Path("word_sample.docx"),
    Path("lorem_ipsum.docx"),
    Path("powerpoint_sample.pptx"),
    Path("2305.03393v1-pg9-img.png"),
    Path("2206.01062.pdf"),
    Path("test_01.asciidoc"),
]
```

```python
results = doc_converter.convert_all(input_paths)

for res in results:
    print(f"{res.input.file.name:35} → {res.status}")
```
